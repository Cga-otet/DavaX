```python
#Importing libraries and dependencies
import pandas as pd
```

```python
#Read input csv into dataframes
df_absences = pd.read_csv('absences.csv')
df_excp_absences = pd.read_csv('exceptional_absence.csv')
df_timesheets = pd.read_csv('timesheets.csv')
df_training_attendance = pd.read_csv('training_attendance.csv')
```

```python
#Select unique employee ids from timesheets
df_absences_unique = df_timesheets[['employee_id']].drop_duplicates()
df_absences_unique
```

```python
#Select needed columns from all dataframes
df_absences_selected = df_absences[['employee_id', 'employee_name', '']]
```

```python
df_joined = df_absences_unique.merge(df_absences, on='employee_id', how='left')
df_joined = df_joined.merge(df_excp_absences, on='employee_id', how='left')
df_joined = df_joined.merge(df_timesheets, on='employee_id', how='left')
df_joined = df_joined.merge(df_training_attendance, on='employee_id', how='left')
df_joined = df_joined.drop_duplicates()
df_joined
```
